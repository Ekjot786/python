# **student performance**

```python
#Import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('StudentsPerformance.csv')

# Calculate average scores by gender
average_scores_by_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
print("Average scores by gender:")
print(average_scores_by_gender)

# Calculate overall average scores
overall_avg_scores = df[['math score', 'reading score', 'writing score']].mean()
print("\nOverall average scores:")
print(overall_avg_scores)

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Scores')
plt.show()

# Countplot for parental level of education
plt.figure(figsize=(10,6))
sns.countplot(y='parental level of education', data=df, order=df['parental level of education'].value_counts().index)
plt.title('Count of Students by Parental Level of Education')
plt.show()

# Pairplot for scores
sns.pairplot(df[['math score', 'reading score', 'writing score']])
plt.suptitle('Pairplot of Scores', y=1.02)
plt.show()

# Previous graphs
# Bar plot
average_scores_by_gender.plot(kind='bar')
plt.title('Bar Plot: Average Scores by Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()

# Line plot
average_scores_by_gender.plot(kind='line', marker='o')
plt.title('Line Plot: Average Scores by Gender')
plt.ylabel('Average Score')
plt.show()

# Box plot for math score by gender
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Box Plot: Math Score by Gender')
plt.show()

# Violin plot for writing score by gender
sns.violinplot(x='gender', y='writing score', data=df)
plt.title('Violin Plot: Writing Score by Gender')
plt.show()

# Interpretation:
# These additional graphs give insights into correlations between scores,
# the distribution of parental education levels, and pairwise relationships
# between math, reading, and writing scores.
```
