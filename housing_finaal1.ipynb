```python
prices = df['SalePrice'].to_numpy()
print("Average House Price:", np.mean(prices))
```

```python
from sklearn.linear_model import LinearRegression

# Features and target
X = df[['LotArea']]
y = df['SalePrice']

# Model
model = LinearRegression()
model.fit(X, y)

# Predict
df['Predicted_Price'] = model.predict(X)

# Compare actual vs predicted
plt.figure(figsize=(8,5))
plt.scatter(df['LotArea'], df['SalePrice'], color='blue', label='Actual Price')
plt.plot(df['LotArea'], df['Predicted_Price'], color='red', label='Predicted Line')
plt.title("Linear Regression: Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend()
plt.show()
```
